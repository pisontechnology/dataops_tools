```python
import pandas as pd
import matplotlib.pyplot as plt
from pyxdf import load_xdf
import numpy as np
import neurokit2 as nk

# Set default figure size for all plots
plt.rcParams['figure.figsize'] = (16, 6)

def preprocess_stream(stream, file_name):
    """
    Preprocesses an XDF stream to extract metadata, timestamps, and data.
    """
    name = stream['info']['name'][0]
    nominal_rate = float(stream['info']['nominal_srate'][0])
    timestamps = np.array(stream['time_stamps'])
    data = np.array(stream['time_series'])

    if len(timestamps) > 1:
        effective_rate = 1 / np.diff(timestamps).mean()
    else:
        effective_rate = 0

    percentage = (effective_rate / nominal_rate * 100) if nominal_rate > 0 else None

    stream_metadata = {
        'Stream Name': name,
        'File Name': file_name,
        'Nominal Rate (Hz)': nominal_rate,
        'Effective Rate (Hz)': effective_rate,
        'Percentage (%)': percentage,
    }

    return {
        "metadata": stream_metadata,
        "timestamps": timestamps,
        "data": data,
        "info": stream['info'],
    }

def get_channel_labels_and_units(stream_info, stream_name, num_channels):
    """
    Extracts channel labels and units from the stream's metadata.
    If labels are not found, assigns default labels.
    """
    labels_and_units = []

    # Attempt to extract labels and units from various possible nesting structures
    if 'desc' in stream_info:
        desc = stream_info['desc']
        if isinstance(desc, list) and len(desc) > 0:
            desc = desc[0]
        if desc and isinstance(desc, dict):
            # For OpenSignals
            if 'channels' in desc:
                channels = desc['channels']
                if isinstance(channels, list) and len(channels) > 0:
                    channels = channels[0]
                if channels and isinstance(channels, dict):
                    channel_list = channels.get('channel', [])
                    if isinstance(channel_list, list):
                        for channel in channel_list:
                            label = channel.get('label', [None])[0] if 'label' in channel else None
                            unit = channel.get('unit', [None])[0] if 'unit' in channel else None
                            labels_and_units.append((label, unit))
            # For Pison Data
            elif 'channel_names' in desc:
                labels = desc['channel_names']
                units = desc.get('unit', [None]*len(labels))
                labels_and_units = list(zip(labels, units))
            else:
                print(f"Warning: Could not find channel labels in 'desc' for stream '{stream_name}'.")
        else:
            print(f"Warning: 'desc' is None or not a dict for stream '{stream_name}'.")
    else:
        print(f"Warning: No 'desc' found in stream '{stream_name}'.")

    # If labels are not found, create default labels
    if not labels_and_units:
        labels_and_units = [(f"Channel {i}", None) for i in range(num_channels)]

    return labels_and_units

def process_and_plot_signal(label, data, sampling_rate, signal_type=None, time=None):
    """
    Processes and plots a signal based on its type using NeuroKit2.
    For audio signals, plots both raw waveform vs sample index and vs timestamp.
    """
    if sampling_rate <= 0 or data.size == 0:
        print(f"Invalid {label} data or sampling rate ({sampling_rate}). Skipping...")
        return None, None

    print(f"\nProcessing {label}: Sampling Rate = {sampling_rate} Hz, Data Size = {data.size}, Signal Type = {signal_type}")

    processed_signals = None
    info = None

    try:
        if signal_type == 'ecg':
            print(f"--> Processing ECG signal for {label}")
            processed_signals, info = nk.ecg_process(data, sampling_rate=sampling_rate)
            nk.ecg_plot(processed_signals, info)
            plt.show()
        elif signal_type == 'ppg':
            print(f"--> Processing PPG signal for {label}")
            processed_signals, info = nk.ppg_process(data, sampling_rate=sampling_rate)
            nk.ppg_plot(processed_signals, info)
            plt.show()
        elif signal_type == 'eda':
            print(f"--> Processing EDA signal for {label}")
            processed_signals, info = nk.eda_process(data, sampling_rate=sampling_rate)
            nk.eda_plot(processed_signals, info)
            plt.show()
        elif signal_type == 'emg':
            print(f"--> Processing EMG signal for {label}")
            processed_signals, info = nk.emg_process(data, sampling_rate=sampling_rate)
            nk.emg_plot(processed_signals, info)
            plt.show()
        elif signal_type == 'audio':
            print(f"--> Plotting Audio signal for {label}")
            if time is not None:
                # Plot Audio vs Sample Index
                plt.figure()
                sample_indices = np.arange(len(data))
                if data.ndim > 1:
                    for channel_idx in range(data.shape[1]):
                        plt.plot(sample_indices, data[:, channel_idx], label=f"{label} - Channel {channel_idx}")
                else:
                    plt.plot(sample_indices, data, label=label)
                plt.title(f"Audio Waveform vs Sample Index for {label}")
                plt.xlabel("Sample Index")
                plt.ylabel("Amplitude")
                plt.legend()
                plt.grid()
                plt.show()

                # Plot Audio vs Time
                plt.figure()
                if data.ndim > 1:
                    for channel_idx in range(data.shape[1]):
                        plt.plot(time, data[:, channel_idx], label=f"{label} - Channel {channel_idx}")
                else:
                    plt.plot(time, data, label=label)
                plt.title(f"Audio Waveform vs Time for {label}")
                plt.xlabel("Time (s)")
                plt.ylabel("Amplitude")
                plt.legend()
                plt.grid()
                plt.show()
            else:
                # If time data is not provided, fallback to plotting against sample index only
                plt.figure()
                sample_indices = np.arange(len(data))
                if data.ndim > 1:
                    for channel_idx in range(data.shape[1]):
                        plt.plot(sample_indices, data[:, channel_idx], label=f"{label} - Channel {channel_idx}")
                else:
                    plt.plot(sample_indices, data, label=label)
                plt.title(f"Audio Waveform vs Sample Index for {label}")
                plt.xlabel("Sample Index")
                plt.ylabel("Amplitude")
                plt.legend()
                plt.grid()
                plt.show()
        else:
            # If signal_type is not specified, try to infer from label
            label_lower = label.lower().strip()
            if "ecg" in label_lower:
                print(f"--> Processing ECG signal for {label}")
                processed_signals, info = nk.ecg_process(data, sampling_rate=sampling_rate)
                nk.ecg_plot(processed_signals, info)
                plt.show()
            elif "ppg" in label_lower:
                print(f"--> Processing PPG signal for {label}")
                processed_signals, info = nk.ppg_process(data, sampling_rate=sampling_rate)
                nk.ppg_plot(processed_signals, info)
                plt.show()
            elif "eda" in label_lower:
                print(f"--> Processing EDA signal for {label}")
                processed_signals, info = nk.eda_process(data, sampling_rate=sampling_rate)
                nk.eda_plot(processed_signals, info)
                plt.show()
            elif "emg" in label_lower:
                print(f"--> Processing EMG signal for {label}")
                processed_signals, info = nk.emg_process(data, sampling_rate=sampling_rate)
                nk.emg_plot(processed_signals, info)
                plt.show()
            elif "audio" in label_lower or "sound" in label_lower or "mic" in label_lower:
                print(f"--> Plotting Audio signal for {label}")
                if time is not None:
                    # Plot Audio vs Sample Index
                    plt.figure()
                    sample_indices = np.arange(len(data))
                    if data.ndim > 1:
                        for channel_idx in range(data.shape[1]):
                            plt.plot(sample_indices, data[:, channel_idx], label=f"{label} - Channel {channel_idx}")
                    else:
                        plt.plot(sample_indices, data, label=label)
                    plt.title(f"Audio Waveform vs Sample Index for {label}")
                    plt.xlabel("Sample Index")
                    plt.ylabel("Amplitude")
                    plt.legend()
                    plt.grid()
                    plt.show()

                    # Plot Audio vs Time
                    plt.figure()
                    if data.ndim > 1:
                        for channel_idx in range(data.shape[1]):
                            plt.plot(time, data[:, channel_idx], label=f"{label} - Channel {channel_idx}")
                    else:
                        plt.plot(time, data, label=label)
                    plt.title(f"Audio Waveform vs Time for {label}")
                    plt.xlabel("Time (s)")
                    plt.ylabel("Amplitude")
                    plt.legend()
                    plt.grid()
                    plt.show()
                else:
                    # If time data is not provided, fallback to plotting against sample index only
                    plt.figure()
                    sample_indices = np.arange(len(data))
                    if data.ndim > 1:
                        for channel_idx in range(data.shape[1]):
                            plt.plot(sample_indices, data[:, channel_idx], label=f"{label} - Channel {channel_idx}")
                    else:
                        plt.plot(sample_indices, data, label=label)
                    plt.title(f"Audio Waveform vs Sample Index for {label}")
                    plt.xlabel("Sample Index")
                    plt.ylabel("Amplitude")
                    plt.legend()
                    plt.grid()
                    plt.show()
            else:
                print(f"--> No processing function available for {label}. Skipping processing.")
                return None, None
    except Exception as e:
        print(f"--> Error processing {label}: {e}")
        return None, None

    return processed_signals, info

def plot_stream(stream_data):
    """
    Plots the time series data for a given stream.
    """
    name = stream_data["metadata"]["Stream Name"]
    timestamps = stream_data["timestamps"]
    data = stream_data["data"]
    info = stream_data["info"]
    sampling_rate = stream_data["metadata"]["Effective Rate (Hz)"]
    time_since_start = timestamps - timestamps[0]

    num_channels = data.shape[1] if data.ndim > 1 else 1
    labels_and_units = get_channel_labels_and_units(info, name, num_channels)

    print(f"\nLabels for stream '{name}': {[label for label, _ in labels_and_units]}")

    processed_signals_dict = {}

    # Determine signal type based on stream name if labels are generic
    stream_name_lower = name.lower()
    if "emg" in stream_name_lower:
        signal_type = "emg"
    elif "ppg" in stream_name_lower:
        signal_type = "ppg"
    elif "ecg" in stream_name_lower:
        signal_type = "ecg"
    elif "audio" in stream_name_lower or "sound" in stream_name_lower or "mic" in stream_name_lower:
        signal_type = "audio"
    else:
        signal_type = None  # Default to None if signal type is not identified

    print(f"Determined signal_type for stream '{name}': {signal_type}")

    # Process signals for known types and store processed signals
    for i in range(num_channels):
        label, unit = labels_and_units[i] if i < len(labels_and_units) else (f"Channel {i}", None)
        if name == "OpenSignals" and label == "nSeq":
            print(f"--> Skipping 'nSeq' channel.")
            continue  # Skip 'nSeq' channel
        channel_data = data[:, i] if data.ndim > 1 else data

        # Determine signal type based on stream name or label
        if signal_type is not None:
            signal_type_channel = signal_type
        else:
            label_lower = label.lower().strip()
            if "emg" in label_lower:
                signal_type_channel = "emg"
            elif "ppg" in label_lower:
                signal_type_channel = "ppg"
            elif "ecg" in label_lower:
                signal_type_channel = "ecg"
            elif "eda" in label_lower:
                signal_type_channel = "eda"
            elif "audio" in label_lower or "sound" in label_lower or "mic" in label_lower:
                signal_type_channel = "audio"
            else:
                signal_type_channel = None

        print(f"Channel {i}: label='{label}', signal_type_channel='{signal_type_channel}'")

        processed_signals, info = process_and_plot_signal(
            label, channel_data, sampling_rate, signal_type=signal_type_channel, time=time_since_start if signal_type_channel == "audio" else None
        )
        processed_signals_dict[label] = processed_signals

    # Plot all channels in the same chart (excluding audio streams)
    if signal_type != "audio":
        plt.figure()
        if name == "OpenSignals":
            # Plotting logic for OpenSignals remains unchanged
            ax1 = plt.gca()
            ax2 = ax1.twinx()
            for i in range(num_channels):
                label, unit = labels_and_units[i] if i < len(labels_and_units) else (f"Channel {i}", None)
                if label == "nSeq":
                    continue  # Skip 'nSeq' channel
                channel_data = data[:, i] if data.ndim > 1 else data
                legend_label = f"{label} ({unit})" if unit else label
                if label == "ECG0":
                    ax2.plot(time_since_start, channel_data, label=legend_label, color='tab:red')
                    ax2.set_ylabel('ECG0', color='tab:red')
                    ax2.tick_params(axis='y', labelcolor='tab:red')
                else:
                    ax1.plot(time_since_start, channel_data, label=legend_label)
            ax1.set_xlabel("Time Since Start (s)")
            ax1.set_ylabel("Signal Value")
            lines_1, labels_1 = ax1.get_legend_handles_labels()
            lines_2, labels_2 = ax2.get_legend_handles_labels()
            ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')
            ax1.grid()
            plt.title(f"Time Series Plot for {name}")
            plt.show()
        else:
            # For Pison Data and other streams
            plt.figure()
            for i in range(num_channels):
                label, unit = labels_and_units[i] if i < len(labels_and_units) else (f"Channel {i}", None)
                channel_data = data[:, i] if data.ndim > 1 else data

                processed_signals = processed_signals_dict.get(label)

                if processed_signals is not None:
                    # Plot processed signals if available
                    if 'PPG_Clean' in processed_signals:
                        plt.plot(time_since_start, processed_signals['PPG_Clean'], label=f"{label} (Processed)")
                    elif 'EMG_Clean' in processed_signals:
                        plt.plot(time_since_start, processed_signals['EMG_Clean'], label=f"{label} (Processed)")
                    else:
                        # If no processed signal available, plot raw data
                        plt.plot(time_since_start, channel_data, label=f"{label} (Raw)")
                else:
                    plt.plot(time_since_start, channel_data, label=f"{label} (Raw)")

            plt.title(f"Time Series Plot for {name}")
            plt.xlabel("Time Since Start (s)")
            plt.ylabel("Signal Value")
            plt.legend()
            plt.grid()
            plt.show()
    else:
        # For Audio streams, plotting was already handled in process_and_plot_signal
        pass

def main():
    """
    Main function to load XDF file, process streams, and display metadata.
    """
    xdf_file_path = "sub-P001_ses-S001_task-Default_run-001_eeg (1).xdf"  # Replace with your actual file path
    print(f"Loading XDF file from: {xdf_file_path}")
    streams, header = load_xdf(xdf_file_path)
    print(f"Number of streams found: {len(streams)}")

    stream_info = []
    time_series_data = {}

    for stream in streams:
        preprocessed = preprocess_stream(stream, xdf_file_path.split('/')[-1])
        stream_info.append(preprocessed["metadata"])
        time_series_data[preprocessed["metadata"]["Stream Name"]] = preprocessed

    stream_info_df = pd.DataFrame(stream_info)
    print("\nStream Information:")
    print(stream_info_df)

    for stream_name, stream_data in time_series_data.items():
        plot_stream(stream_data)

    # Display the DataFrame (works in Jupyter notebooks)
    try:
        from IPython.display import display
        display(stream_info_df)
    except ImportError:
        print("\nStream Information DataFrame:")
        print(stream_info_df)

if __name__ == "__main__":
    main()
```
